### 7. Pub/Sub Messaging

Pub/Sub messaging enables service to service communication asynchronously
- Messaging in Pub/Sub context is different from email or text messages, because it is intended for communication between software services, not between people
- A message typically in small size (in the tens or hundreds of KBs) contains the metadata of the application or service 
- A messaging server or service allows a place for your messages to be stored temporarily and transmitted
- In a messaging service, the two main resources are the queue and the topic

A queue is like a buffer. You can put messages into a queue, and you can retrieve messages from a queue. The service  that puts messages into a queue is called a message producer and the service that retrieves messages is called a message consumer




A topic is like a multicast in Networking, like a OSPF neighbors are dynamically discovered by sending Hello packets out to the multicast IP address of 224.0.0.5 and the other OSPF neighbors who's interested in the same  multicast IP address of 224.0.0.5 start communicating further on. In simialr way, a service can publish messages to a topic, and anyone interested in these messages can subscribe to the topic. Then, the interested parties are notified about the published messages. The service that broadcasts topics is called a topic publisher and the service that subscribes to broadcasts is called a topic subscriber.




#### Messaging service examples:
**Relatime system and application events and logs is one of the most common use case of messaging service:** Imagine network systems sending syslog messages to its single server or to master server (in HA model), if any other system interested in all or some of those messages needs to talk to either to syslog server or configure 2nd syslog server in each of those clients. To configure one more and sending the same logs to second system is not bad. But imagine, if you have 100k of clients streamsing the syslog messages and 1000s of different systems or services interested in all or some of those logs ? This is when pub/sub messaging service could help solve the scalability problem. All those 100k+ clients can stream or publish thier logs to messaging service and other services can subscribe to the messages they are interested in.
